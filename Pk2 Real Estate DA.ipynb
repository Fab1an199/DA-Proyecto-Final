```python
def missing_cols(df):
    total = 0
    for col in df.columns:
        missing_vals = df[col].isnull().sum()
        total += missing_vals
        if missing_vals != 0:
            print(f"{col} =>{df[col].isnull().sum()}")
    if total == 0:
        print ("No hay valores faltantes")
missing_cols(dfpropiedades)
```

```python
average_sale_amount = df.groupby('property_type')['sale_amount'].mean().reset_index()
average_sale_amount = average_sale_amount.sort_values(by='sale_amount')
print(average_sale_amount)
```

```python
fig, ax = plt.subplots(1,1, figsize = (15,6.9), dpi = 100)

colores = ['#33384e', '#33384e','#33384e','#33384e','#33384e','#33384e','#33384e','#33384e',
          '#33384e','#33384e','#33384e','#cc2649']
        
ax.bar(average_sale_amount['property_type'], average_sale_amount['sale_amount'], color=colores)

plt.ylabel('Costo Promedio')

titulo = 'Grafico 4: Costo Promedio de las propiedades residenciales y comerciales'
plt.suptitle(titulo,
             horizontalalignment = 'left',
            x = 0.001,
            y = 0.95,
            transform = fig.transFigure,
            fontsize = 20,
            fontweight = 'bold',
            fontfamily = 'arial')

etiquetaabajo = 'Tipo de Propiedad'
ax.set_xlabel(etiquetaabajo,
             ha = 'left',
             fontsize = 10,
             fontweight = 'regular',
             fontfamily = 'arial')
ax.set_ylabel('Costo Promedio')

ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, _: f'{int(y / 1e6)}M'))

plt.savefig('Grafico4.png')

plt.show()
```
